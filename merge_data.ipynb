```python
import pandas as pd
```

```python
def get_merge_df(cv_path: str, gowajee_path: str, smart_home_path: str, health_path: str) -> pd.DataFrame:
    cv_df = pd.read_csv(cv_path)
    cv_df["path"] = "/mnt/d/data/cv-corpus-13.0-2023-03-09/th/clips/" + cv_df["path"]

    gowajee_df = pd.read_csv(gowajee_path)
    gowajee_df["path"] = "/mnt/d/data/gowajee_v0.9.2/v0.9.2/" + gowajee_df["path"]
    
    smart_home_df = pd.read_csv(smart_home_path)
    smart_home_df["path"] = "/mnt/d/data/Dataset/Smarthome/Record/" + smart_home_df["path"]
    
    health_df = pd.read_csv(health_path)
    health_df["path"] = "/mnt/d/data/Dataset/Healthcare/Record/" + health_df["path"]

    df = pd.concat([cv_df, gowajee_df, smart_home_df, health_df])
    return df
```

```python
train = get_merge_df(cv_path="train.csv", gowajee_path="gowajee_train.csv", smart_home_path="smart_home_train.csv", health_path="health_train.csv")
dev = get_merge_df(cv_path="dev.csv", gowajee_path="gowajee_dev.csv", smart_home_path="smart_home_dev.csv", health_path="health_dev.csv")
test = get_merge_df(cv_path="test.csv", gowajee_path="gowajee_test.csv", smart_home_path="smart_home_test.csv", health_path="health_test.csv")
```

```python
train.to_csv("train_merge.csv", index=False)
dev.to_csv("dev_merge.csv", index=False)
test.to_csv("test_merge.csv", index=False)
```
